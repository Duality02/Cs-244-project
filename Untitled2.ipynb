```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import time

# Take the search link for a product and loop in the search page for all the links about the product that it can find, then go to the next page until all the pages are done
next_page = ""
all_data = []
ct = 0
url = "https://www.jumia.com.eg/catalog/?q=hp"
i = 0
while ct < 50:
    ct += 1
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    html_content = response.content
    soup = BeautifulSoup(html_content, "html.parser")
    
    # Find the next page link
    link_element = soup.find("a", class_="pg", href=True)
    next_page = link_element.get("href") if link_element else None

    head_link = "https://www.jumia.com.eg/"
    links = soup.find("div", class_="-paxs row _no-g _4cl-3cm-shs").find_all("a")
    for link in links:
        i += 1
        page_content = requests.get(head_link + link["href"], headers=headers).content
        soup = BeautifulSoup(page_content, "html.parser")
        product_name_element = soup.find("h1", class_="-fs20 -pts -pbxs")
        product_name = product_name_element.text.strip() if product_name_element else "N/A"
        price_range_element = soup.find("span", class_="-b -ltr -tal -fs24")
        price_range = price_range_element.text.strip() if price_range_element else "N/A"
        ratings_element = soup.find("div", class_="-df -i-ctr -pvxs")
        ratings = ratings_element.text.strip() if ratings_element else "N/A"
        company_name_element = soup.find("div", class_="row card _no-g -fg1 -pas")
        company_name = company_name_element.find("div", class_="col10").find("div", class_="-pvxs").text.split() if company_name_element else "N/A"
        product_image_element = soup.find("div", class_="-ptxs -pbs")
        product_image = product_image_element.find("img").get("data-src") if product_image_element else "N/A"
        if company_name != "N/A":
            all_data.append([head_link + link["href"], product_name, price_range, ratings, company_name, product_image])
        if i >= 199:
            time.sleep(60)
            i = 0
    url = head_link + next_page

filename = "hp.csv"
header = ["Product Page", "Product Name", "Price Range", "Ratings", "Company Name", "Product Image"]

with open(filename, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(header)
    for data in all_data:
        writer.writerow(data)
print("Data has been saved successfully.")

df = pd.read_csv(filename)
df
```
